```python
import math

import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100 #NUMBER OF TIME STEPS
T = 2.
dt = T/100.
X0 = 90 #INITIAL STOCK PRICE
num = 10000 #NUMBER OF SIMULATION
a = 0.1 #MEAN REVERSION SPEED
theta = 86 #LONG TIME AVERAGE
sigma = 0.1 #VOLATILTY


def dynamics(a,theta,X_t,dt,sigma):
    dWt = np.random.normal(loc=0,scale=np.sqrt(dt))
    dXt = a*(theta-X_t)*dt + sigma*np.sqrt(X_t)*dWt
    return dXt

t = np.linspace(0,T,N+1)
X = np.zeros((num,N+1))
X[:,0] = X0
```

```python
E=0
for i in range(num):
    for j in range(1,N+1):
        X[i,j] = X[i,j-1]+dynamics(a,theta,X[i,j-1],dt,sigma)
    E = E + X[i,N]
E = E/(num)
E 
```

```python
plt.figure(figsize=(10, 6))
for i in range(num):
    plt.plot(t, X[i], label=f'Simulation {i+1}')

plt.title('Monte Carlo Simulation of Stock Prices')
plt.xlabel('Time (Years)')
plt.ylabel('Stock Price')
# plt.legend()
# plt.grid(True)
plt.show()
```
